```python
import matplotlib.pyplot as plt

# Assuming 'bb' is your DataFrame

# Check if 'bb' DataFrame is empty or if 'city' column exists
if bb.empty or 'city' not in bb.columns:
    print("DataFrame is empty or 'city' column is missing. Cannot create bar chart.")
else:
    # Example 3: Bar chart of 'city' counts
    city_counts = bb['city'].value_counts()
    plt.figure(figsize=(8, 5))
    city_counts.plot(kind='bar')
    plt.xlabel('City')
    plt.ylabel('Count')
    plt.title('Bar Chart of City Counts')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
    plt.tight_layout()
    plt.show()
```
